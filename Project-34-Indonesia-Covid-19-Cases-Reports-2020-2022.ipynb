```python
# Simpan salinan data awal untuk perbandingan
df_before = df.copy()

# --- Deteksi dan Hapus Duplikat ---
print("Jumlah data duplikat sebelum cleaning:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Jumlah data duplikat setelah cleaning:", df.duplicated().sum())
```

```python
# --- Deteksi dan Hapus Missing Values ---
print("\nMissing values sebelum cleaning:")
print(df.isnull().sum())
# Misal: kita hapus seluruh baris yang memiliki missing value
df.dropna(inplace=True)
print("\nMissing values setelah cleaning:")
print(df.isnull().sum())

# --- Deteksi dan Hapus Outliers ---
# Fungsi untuk menghapus outlier dengan metode IQR untuk kolom numerik
def remove_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    # Batas bawah dan atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Filter data
    return data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
```

```python
# Misal kita lakukan pembersihan outlier untuk beberapa kolom kunci
cols_to_clean = ['New_Cases', 'New_Deaths', 'New_Recovered']
for col in cols_to_clean:
    df = remove_outliers(df, col)
```

```python
# --- Visualisasi Histogram dan Box Plot ---

# Fungsi untuk menggambar histogram dan box plot untuk kolom tertentu
def plot_comparison(df_before, df_after, col):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    # Histogram sebelum cleaning
    sns.histplot(df_before[col], ax=axes[0,0], kde=True)
    axes[0,0].set_title(f'Histogram {col} (Sebelum Cleaning)')
    # Boxplot sebelum cleaning
    sns.boxplot(x=df_before[col], ax=axes[0,1])
    axes[0,1].set_title(f'Box Plot {col} (Sebelum Cleaning)')
    # Histogram setelah cleaning
    sns.histplot(df_after[col], ax=axes[1,0], kde=True)
    axes[1,0].set_title(f'Histogram {col} (Setelah Cleaning)')
    # Boxplot setelah cleaning
    sns.boxplot(x=df_after[col], ax=axes[1,1])
    axes[1,1].set_title(f'Box Plot {col} (Setelah Cleaning)')
    plt.tight_layout()
    plt.show()
```

```python
# Visualisasi untuk kolom New_Cases
plot_comparison(df_before, df, 'New_Cases')
```
